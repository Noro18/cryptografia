```python
# CARSAR CYPHER

def encrypt(key, plaintext):
    plaintext = plaintext.upper()
    alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    result = ""


    for letter in plaintext:
        if letter in alpha:
            letter_index = (alpha.find(letter) + key) % len(alpha)

            result += alpha[letter_index]
        else:
            result += letter

    return result
            

def decrypt(key, ciphertext):
    cipher = ciphertext.upper()
    alpha =  "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    result = ""

    for letter in cipher:
        if letter in alpha:
            letter_index = (alpha.find(letter) - key) % len(alpha)
            result += alpha[letter_index]
        else:
            result += letter

    return result


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import string

def frequency_analysis(text):
    # 1. Normalize text
    text = text.lower()

    # Keep only alphabet letters
    text = [char for char in text if char in string.ascii_lowercase]

    # 2. Count frequencies
    freq_series = pd.Series(text).value_counts().sort_index()

    # 3. Ensure all letters appear (even if 0)
    all_letters = pd.Series(0, index=list(string.ascii_lowercase))
    freq_series = all_letters.add(freq_series, fill_value=0)

    # 4. Convert to DataFrame for clarity
    df = pd.DataFrame({
        'letter': freq_series.index,
        'frequency': freq_series.values
    })

    return df


def plot_frequency(df):
    # Sort by frequency (descending)
    df_sorted = df.sort_values(by='frequency', ascending=False)

    plt.figure()

    plt.bar(df_sorted['letter'], df_sorted['frequency'])

    plt.title("Letter Frequency Analysis (Sorted)")
    plt.xlabel("Letters")
    plt.ylabel("Frequency")

    plt.grid(axis='y')

    plt.show()


if __name__ == "__main__":
    text_input = input("Enter your text: ")

    df = frequency_analysis(text_input)

    print("\nFrequency Table:")
    print(df)

    plot_frequency(df)
```
