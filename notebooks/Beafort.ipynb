```python
def beafort_encrypt(text, key):
    plain_text = text
    key = key.upper()

    # remove spaces for key alignment
    clean_plain = plain_text.replace(" ", "").upper()

    # repeat key
    repeated_key = (key * ((len(clean_plain) // len(key)) + 1))[:len(clean_plain)]

    letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    letter_to_number = {letter: index for index, letter in enumerate(letters)}
    number_to_letter = {index: letter for index, letter in enumerate(letters)}

    result = []
    key_index = 0
    i = 1

    for letter in plain_text.upper():
        if letter == " ":
            result.append(" ")
            print("added space")
        else:
            k = repeated_key[key_index]

            number = (letter_to_number[k] - letter_to_number[letter]) % 26

            print("========================")
            print(f"C{i} = {k} - {letter} % 26")
            print(f"C{i} = {letter_to_number[k]} - {letter_to_number[letter]} = {number} = {number_to_letter[number]}")

            result.append(number_to_letter[number])

            key_index += 1
            i += 1

    return "".join(result)

def beafort_decrypt(text, key):
    plain_text = text
    key = key.upper()

    # remove spaces for key alignment
    clean_plain = plain_text.replace(" ", "").upper()

    # repeat key
    repeated_key = (key * ((len(clean_plain) // len(key)) + 1))[:len(clean_plain)]

    letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    letter_to_number = {letter: index for index, letter in enumerate(letters)}
    number_to_letter = {index: letter for index, letter in enumerate(letters)}

    result = []
    key_index = 0
    i = 1


    print("")
    print("")
    print("================Decrypting===============")
    for letter in plain_text.upper():
        if letter == " ":
            result.append(" ")
            print("added space")
        else:
            k = repeated_key[key_index]

            number = (letter_to_number[k] - letter_to_number[letter]) % 26

            print("========================")
            print(f"D{i} = {k} - {letter} % 26")
            print(f"D{i} = {letter_to_number[k]} - {letter_to_number[letter]} % 26 = {number} = {number_to_letter[number]}")

            result.append(number_to_letter[number])

            key_index += 1
            i += 1

    return "".join(result)
```

```python

plain_text = "MY NAME IS XISTO NORONHA SOARES DA CUNHA FROM MANATUTO DISTRICT"
key = "STUDENT"

ciphertext = beafort_encrypt(plain_text, key)
print(ciphertext)
deciphertext = beafort_decrypt(ciphertext, key)
print(deciphertext)
```
