# Vigenere Text 


````python
plain_text = "THIS PLAINTEXT IS SECRET"
key = "MAYBE"

# remove spaces
clean_plain = plain_text.replace(" ", "")

# repeat key to match length of text-without-spaces
repeated_key = (key * ((len(clean_plain) // len(key)) + 1))[:len(clean_plain)]

print(repeated_key)  # 'MAYB EMAYB EMAYB EMAYB EM'

letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# create a dictionary to map letters to numbers
letter_to_number = {letter: index for index, letter in enumerate(letters)}
print(letter_to_number)

number_to_letter = {index: letter for index, letter in enumerate(letters)}



````

````python


result = []

for letter, k in zip(plain_text, repeated_key):
    if letter == " ":
        result.append(" ")
    else:
        number = (letter_to_number[key] + letter_to_number[letter]) % 26
        result.append(number_to_letter[number])

cipher_text  = "".join(result)

print(cipher_text)
````

````python


result = []

for letter, k in zip(cipher_text, repeated_key):
    if letter == " ":
        result.append(" ")
    else:
        number = (letter_to_number[letter] - letter_to_number[k]) % 26
        result.append(number_to_letter[number])

print("".join(result))



````

````python
import collections
import matplotlib.pyplot as plt

cipher = "VEVZEVXEVCEEVZEVMEDEVNEVLEEVBEVSEVREVEEVFE"

# count only letters
counts = collections.Counter(ch for ch in cipher if ch.isalpha())

letters = sorted(counts.keys())
frequencies = [counts[ch] for ch in letters]

print("Counts:", counts)

plt.figure(figsize=(8, 4))
plt.bar(letters, frequencies, color="skyblue", edgecolor="black")
plt.xlabel("Letter")
plt.ylabel("Frequency")
plt.title("Letter Frequency in Ciphertext")
plt.grid(axis="y", linestyle="--", alpha=0.3)
plt.show()
````
